```python
# Installation des dépendances
!pip install scikit-learn imbalanced-learn nltk nlpaug matplotlib seaborn wordcloud

# Importation des bibliothèques
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import nltk
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize
from nltk.stem import WordNetLemmatizer
import re
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics import confusion_matrix
from collections import Counter
from google.colab import drive
from wordcloud import WordCloud
import time

nltk.download('punkt', quiet=True)
nltk.download('stopwords', quiet=True)
nltk.download('wordnet', quiet=True)
nltk.download('omw-1.4', quiet=True)
nltk.download('averaged_perceptron_tagger', quiet=True)
nltk.download('punkt_tab', quiet=True) # Normalement couvert par 'punkt'

drive.mount('/content/gdrive/')
notebook_path = '/content/gdrive/My Drive/Colab Notebooks/'
%cd "$notebook_path"

print("--- CHARGEMENT ET PRÉPARATION DES DONNÉES POUR ANALYSE (SECTION 2.4) ---")
file_path_final_annotations = 'Annotation_apres_mise_en_accord_a_utiliser_pour_entrainement_modele.tsv'
try:
    df_final_annotated = pd.read_csv(file_path_final_annotations, sep='\t')
except FileNotFoundError:
    print(f"ERREUR: Le fichier d'annotations finales '{file_path_final_annotations}' est introuvable.")
    raise
print(f"Nombre total de tweets : {len(df_final_annotated)}")
label_column_final = 'emotion_principale'
if label_column_final not in df_final_annotated.columns:
    raise ValueError(f"Colonne des labels '{label_column_final}' introuvable.")
print(f"Colonne des labels : {label_column_final}")

stop_words_for_analysis = set(stopwords.words('english'))
stop_words_for_analysis.update(['url', 'http', 'https', 'www', 'co', 'rt', 'amp', 'via'])
lemmatizer_for_analysis = WordNetLemmatizer()

def clean_text_for_analysis(text):
    if not isinstance(text, str): return ""
    text = text.lower()
    text = re.sub(r'http\S+|www\S+|https\S+', '', text, flags=re.MULTILINE)
    text = re.sub(r'@\w+', '', text)
    text = re.sub(r'[^\w\s#]', ' ', text)
    text = re.sub(r'\d+', '', text)
    text = re.sub(r'\s+', ' ', text).strip()
    return text

def process_text_stopwords_lemmatize_for_analysis(text):
    tokens = word_tokenize(text)
    lemmatized_tokens = [lemmatizer_for_analysis.lemmatize(token) for token in tokens]
    filtered_tokens = [token for token in lemmatized_tokens if token not in stop_words_for_analysis and len(token) > 2]
    return ' '.join(filtered_tokens)

df_final_annotated['text_original_for_wc'] = df_final_annotated['text'].astype(str)
df_final_annotated['cleaned_text_for_analysis'] = df_final_annotated['text'].apply(clean_text_for_analysis)
df_final_annotated['processed_text_for_wordclouds_and_tfidf'] = df_final_annotated['cleaned_text_for_analysis'].apply(process_text_stopwords_lemmatize_for_analysis)
df_final_annotated = df_final_annotated[df_final_annotated['cleaned_text_for_analysis'].str.strip() != '']
print(f"Tweets restants après nettoyage : {len(df_final_annotated)}")

if 'science_related' not in df_final_annotated.columns:
    print("AVERTISSEMENT: Colonne 'science_related' non trouvée. Ajout d'une colonne factice.")
    df_final_annotated['science_related'] = 0
df_final_annotated['label_code'] = df_final_annotated[label_column_final] - 1
label_map_analysis = {0: 'Peur', 1: 'Colère', 2: 'Joie', 3: 'Surprise', 4: 'Tristesse', 5: 'Dégoût', 6: 'Neutre'}
df_final_annotated['emotion_name'] = df_final_annotated['label_code'].map(label_map_analysis)

print("\n" + "="*60)
print("ANALYSES POUR LA SECTION 2.4 DU RAPPORT")
print("="*60)

# 2.4.1 Distribution des Émotions
print("\n--- 2.4.1 Distribution des Émotions ---")
emotion_order = df_final_annotated['emotion_name'].value_counts().index
plt.figure(figsize=(10, 6))
ax_global = sns.countplot(data=df_final_annotated, x='emotion_name', order=emotion_order, palette='viridis', hue='emotion_name', legend=False)
plt.title('Distribution Globale des Émotions'); plt.xlabel('Émotion'); plt.ylabel('Nombre de Tweets')
plt.xticks(rotation=45, ha='right');
for container in ax_global.containers: ax_global.bar_label(container, fmt='%.0f')
plt.tight_layout(); plt.show()
print("Distribution globale (%):\n", df_final_annotated['emotion_name'].value_counts(normalize=True).mul(100).round(2).astype(str) + '%')
print("Distribution globale (comptes):\n", df_final_annotated['emotion_name'].value_counts())

df_sci = df_final_annotated[df_final_annotated['science_related'] == 1]
df_non_sci = df_final_annotated[df_final_annotated['science_related'] == 0]
print(f"\nTweets scientifiques: {len(df_sci)}, Non-scientifiques: {len(df_non_sci)}")
if not df_sci.empty or not df_non_sci.empty:
    fig, axes = plt.subplots(1, 2, figsize=(16, 6), sharey=False)
    if not df_sci.empty:
        sns.countplot(data=df_sci, x='emotion_name', order=emotion_order, palette='mako', hue='emotion_name', legend=False, ax=axes[0])
        axes[0].set_title(f'Émotions (Scientifiques, N={len(df_sci)})'); axes[0].set_xlabel('Émotion'); axes[0].set_ylabel('N Tweets')
        axes[0].tick_params(axis='x', rotation=45, labelsize=9); [axes[0].bar_label(c, fmt='%.0f') for c in axes[0].containers]
    else: axes[0].text(0.5,0.5,"Aucun tweet scientifique",ha='center',va='center',transform=axes[0].transAxes); axes[0].set_title('Émotions (Scientifiques)')
    if not df_non_sci.empty:
        sns.countplot(data=df_non_sci, x='emotion_name', order=emotion_order, palette='rocket', hue='emotion_name', legend=False, ax=axes[1])
        axes[1].set_title(f'Émotions (Non-Scientifiques, N={len(df_non_sci)})'); axes[1].set_xlabel('Émotion'); axes[1].set_ylabel('N Tweets')
        axes[1].tick_params(axis='x', rotation=45, labelsize=9); [axes[1].bar_label(c, fmt='%.0f') for c in axes[1].containers]
    else: axes[1].text(0.5,0.5,"Aucun tweet non-scientifique",ha='center',va='center',transform=axes[1].transAxes); axes[1].set_title('Émotions (Non-Scientifiques)')
    plt.tight_layout(); plt.show()
    if not df_sci.empty: print("\nScientifiques (%):\n", df_sci['emotion_name'].value_counts(normalize=True).mul(100).round(2).astype(str) + '%')
    if not df_non_sci.empty: print("Non-Scientifiques (%):\n", df_non_sci['emotion_name'].value_counts(normalize=True).mul(100).round(2).astype(str) + '%')
    print("\n--- Proportion Scientifique/Non-Scientifique par Émotion ---")
    emotion_context_dist = df_final_annotated.groupby('emotion_name')['science_related'].value_counts(normalize=True).mul(100).round(1).unstack(fill_value=0)
    if 0 not in emotion_context_dist.columns: emotion_context_dist[0] = 0.0
    if 1 not in emotion_context_dist.columns: emotion_context_dist[1] = 0.0
    emotion_context_dist = emotion_context_dist.rename(columns={0: 'Non-Scientifique (%)', 1: 'Scientifique (%)'})
    print(emotion_context_dist)
    if not emotion_context_dist.empty:
        try:
            emotion_context_dist[['Scientifique (%)', 'Non-Scientifique (%)']].plot(kind='bar', stacked=True, figsize=(12, 7), colormap='coolwarm_r')
            plt.title('Proportion Tweets Scientifiques vs. Non-Scientifiques par Émotion'); plt.xlabel('Émotion'); plt.ylabel('% Tweets')
            plt.xticks(rotation=45, ha='right'); plt.legend(title='Contexte'); plt.tight_layout(); plt.show()
        except Exception as e: print(f"Erreur graphique barres empilées: {e}")

# 2.4.2 Popularité des "Types de Tweets"
print("\n--- 2.4.2 Popularité Types Tweets (Exemples) ---")
for emotion_code_val, emotion_name_str in label_map_analysis.items():
    print(f"\nExemples tweets pour: {emotion_name_str}")
    df_emotion_sample = df_final_annotated[df_final_annotated['label_code'] == emotion_code_val]
    sample_size = min(3, len(df_emotion_sample))
    if sample_size > 0:
        [print(f"  Ex {i+1}: {t[:150]}...") for i, t in enumerate(df_emotion_sample['text'].sample(sample_size,random_state=42).tolist())]
    else: print(f"  Aucun tweet pour {emotion_name_str}")

# 2.4.3 Visualisations et Statistiques
print("\n--- 2.4.3 Visualisations et Statistiques ---")
print("\nGenerating Wordclouds...")
try:
    text_wc = ' '.join(df_final_annotated['text_original_for_wc'].dropna())
    if text_wc.strip():
        wc = WordCloud(width=1000,height=500,background_color ='white',colormap='magma',max_words=150).generate(text_wc)
        plt.figure(figsize=(12,6));plt.imshow(wc,interpolation='bilinear');plt.axis("off");plt.title("Wordcloud Global (Original Brut)");plt.show()
except Exception as e: print(f"Erreur WC original: {e}")
try:
    text_wc_clean = ' '.join(df_final_annotated['cleaned_text_for_analysis'].dropna())
    if text_wc_clean.strip():
        wc_clean = WordCloud(width=1000,height=500,stopwords=stop_words_for_analysis,background_color ='white',colormap='cividis',max_words=150).generate(text_wc_clean)
        plt.figure(figsize=(12,6));plt.imshow(wc_clean,interpolation='bilinear');plt.axis("off");plt.title("Wordcloud Global (Nettoyé, Stopwords Appliqués)");plt.show()
except Exception as e: print(f"Erreur WC nettoyé: {e}")

for code, name in label_map_analysis.items():
    text_emo = ' '.join(df_final_annotated[df_final_annotated['label_code']==code]['processed_text_for_wordclouds_and_tfidf'].dropna())
    if text_emo.strip() and len(text_emo.split()) > 3:
        try:
            plt.figure(figsize=(10,5));wc_emo = WordCloud(width=800,height=400,background_color='white',colormap='plasma',max_words=100).generate(text_emo)
            plt.imshow(wc_emo,interpolation='bilinear');plt.axis("off");plt.title(f"Wordcloud: {name} (processé)");plt.show()
        except Exception as e: print(f"Erreur WC pour {name}: {e}")
    else: print(f"Pas assez de texte pour WC : {name}")

print("\nMatrice de Confusion Inter-Annotateurs")
path_to_raw_annotations = 'Annotation_sans_mise_en_accord_avec_Ilian.csv'
try:
    df_raw_annotations = pd.read_csv(path_to_raw_annotations, sep=';')
    print(f"Fichier '{path_to_raw_annotations}' chargé.")
    annotator_pairs = {"Alain_vs_Anton": ('Alain_émotions', 'Anton_émotions'),
                       "Alain_vs_Tony": ('Alain_émotions', 'Tony_émotion'),
                       "Anton_vs_Tony": ('Anton_émotions', 'Tony_émotion')}
    for pair_name, (col_A, col_B) in annotator_pairs.items():
        print(f"\n--- Matrice: {pair_name} ---")
        if col_A in df_raw_annotations.columns and col_B in df_raw_annotations.columns:
            df_s = df_raw_annotations[[col_A, col_B]].copy()
            df_s[col_A]=pd.to_numeric(df_s[col_A],errors='coerce'); df_s[col_B]=pd.to_numeric(df_s[col_B],errors='coerce')
            df_s.dropna(subset=[col_A,col_B],inplace=True)
            if not df_s.empty:
                lA=(df_s[col_A].astype(int)-1); lB=(df_s[col_B].astype(int)-1)
                vA=lA.isin(label_map_analysis.keys());vB=lB.isin(label_map_analysis.keys());cVm=vA&vB
                lAf=lA[cVm];lBf=lB[cVm]
                if not lAf.empty:
                    labels_cm=sorted(list(label_map_analysis.keys()));names_cm=[label_map_analysis.get(i,str(i)) for i in labels_cm]
                    cm=confusion_matrix(lAf,lBf,labels=labels_cm)
                    print(f"Matrice ({col_A} vs {col_B}):\nLabels: {names_cm}\n{pd.DataFrame(cm, index=names_cm, columns=names_cm)}")
                    plt.figure(figsize=(9,7));sns.heatmap(cm,annot=True,fmt='d',cmap='coolwarm',xticklabels=names_cm,yticklabels=names_cm)
                    plt.title(f"Confusion: {col_A} vs {col_B}");plt.xlabel(col_B);plt.ylabel(col_A);plt.show()
                else: print(f"Pas de données valides pour {pair_name} après filtrage.")
            else: print(f"Pas de données numériques valides pour {pair_name}.")
        else: print(f"Colonnes pour {pair_name} introuvables.")
except FileNotFoundError: print(f"ERREUR: Fichier '{path_to_raw_annotations}' introuvable.")
except Exception as e: print(f"Erreur matrices de confusion: {e}")

print("\n--- Autres Statistiques Descriptives ---")
df_final_annotated['tweet_length_chars']=df_final_annotated['text_original_for_wc'].apply(len)
df_final_annotated['tweet_length_words']=df_final_annotated['text_original_for_wc'].apply(lambda x: len(x.split()))
print("Longueur moyenne (caractères):\n", df_final_annotated.groupby('emotion_name')['tweet_length_chars'].mean().sort_values(ascending=False).round(1))
print("Longueur moyenne (mots):\n", df_final_annotated.groupby('emotion_name')['tweet_length_words'].mean().sort_values(ascending=False).round(1))

# --- Option 2: Violin Plot avec Limite Y (Décommentez et ajustez y_limit_violin si besoin) ---
# plt.figure(figsize=(13, 7))
# order_emotions_violin = df_final_annotated.groupby('emotion_name')['tweet_length_words'].median().sort_values(ascending=False).index
# sns.violinplot(data=df_final_annotated, x='emotion_name', y='tweet_length_words',
#             order=order_emotions_violin, palette='pastel', hue='emotion_name', legend=False,
#             inner='quartile', cut=0, density_norm='width', linewidth=1.5, dodge=False)
# plt.title('Distribution Longueur Tweets (mots) par Émotion - Violin Plot', fontsize=15)
# plt.xlabel('Émotion', fontsize=12); plt.ylabel('Nombre de Mots', fontsize=12)
# plt.xticks(rotation=45, ha='right', fontsize=10); plt.yticks(fontsize=10)
# y_limit_violin = df_final_annotated['tweet_length_words'].quantile(0.98)
# plt.ylim(0, y_limit_violin + 5)
# plt.grid(axis='y', linestyle='--', alpha=0.7); plt.tight_layout(); plt.show()
# print(f"Note: L'axe Y du Violin Plot a été limité à environ {y_limit_violin:.0f} mots pour une meilleure lisibilité.")


# --- Option 3: Boxplot avec Stripplot (Décommentez si besoin) ---
plt.figure(figsize=(14, 8))
order_emotions_boxstrip = df_final_annotated.groupby('emotion_name')['tweet_length_words'].median().sort_values(ascending=False).index
sns.boxplot(data=df_final_annotated, x='emotion_name', y='tweet_length_words',
            order=order_emotions_boxstrip, palette='pastel', hue='emotion_name', legend=False,
            showfliers=False, linewidth=1.5, boxprops=dict(alpha=.7))
sns.stripplot(data=df_final_annotated, x='emotion_name', y='tweet_length_words',
              order=order_emotions_boxstrip, color='black', alpha=0.25, jitter=0.2, size=3)
plt.title('Distribution Longueur Tweets (mots) par Émotion - Boxplot & Stripplot', fontsize=16)
plt.xlabel('Émotion', fontsize=12); plt.ylabel('Nombre de Mots', fontsize=12)
plt.xticks(rotation=45, ha='right', fontsize=10); plt.yticks(fontsize=10)
plt.grid(axis='y', linestyle='--', alpha=0.7)
y_limit_boxstrip = df_final_annotated['tweet_length_words'].quantile(0.99)
plt.ylim(0, y_limit_boxstrip + 10)
plt.tight_layout(); plt.show()


print("\n--- Mots les plus discriminants par émotion (basé sur TF-IDF moyen) ---")
try:
    vectorizer_discriminant = TfidfVectorizer(max_features=1000, ngram_range=(1,1))
    tfidf_matrix_disc = vectorizer_discriminant.fit_transform(df_final_annotated['processed_text_for_wordclouds_and_tfidf'])
    feature_names_disc = vectorizer_discriminant.get_feature_names_out()

    df_tfidf_disc = pd.DataFrame(tfidf_matrix_disc.toarray(), columns=feature_names_disc, index=df_final_annotated.index)
    df_tfidf_label_disc = pd.concat([df_final_annotated[['label_code', 'emotion_name']], df_tfidf_disc], axis=1)

    N_TOP_WORDS = 7
    for emotion_code_val, emotion_name_val in label_map_analysis.items():
        emotion_data = df_tfidf_label_disc[df_tfidf_label_disc['label_code'] == emotion_code_val]
        if not emotion_data.empty:
            mean_tfidf_emotion = emotion_data[feature_names_disc].mean()
            top_words = mean_tfidf_emotion.sort_values(ascending=False).head(N_TOP_WORDS)
            print(f"\nTop {N_TOP_WORDS} mots (TF-IDF moyen) pour {emotion_name_val}:")
            if not top_words.empty:
                for word, score in top_words.items(): print(f"  - {word}: {score:.4f}")
            else: print(f"  Aucun mot discriminant trouvé pour {emotion_name_val}.")
        else: print(f"\nAucune donnée pour l'émotion {emotion_name_val} dans l'analyse des mots discriminants.")
except Exception as e: print(f"Erreur analyse mots discriminants : {e}")

print("\nFin des analyses pour la section 2.4.")
```
